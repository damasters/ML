```python
acc_list_test_2 = []
acc_list_train_2 = []
precision_list_train_2 = []
precision_list_test_2 = []
recall_list_train_2 = []
recall_list_test_2 = []
f1_list_train_2 = []
f1_list_test_2 = []
auc_list_train_2 = []
auc_list_test_2 = []
for m in range(30):
    print('\nRUN #'+str(m+1))
    semi_train_x, semi_test_x, semi_train_y, semi_test_y = train_test_split(norm_data_x, y_df, test_size = 0.2)
    vals = semi_train_y.values
    #get indexes of values
    pos_vals = {}
    neg_vals = {}
    for i, v in enumerate(vals):
        if v == 1:
            pos_vals[i] = v
        else:
            neg_vals[i] = v
    rand_pos_idx = random.sample(list(pos_vals), len(pos_vals)//2)
    rand_neg_idx = random.sample(list(neg_vals), len(neg_vals)//2)
    all_indexes = rand_pos_idx + rand_neg_idx
    unlab_idx = []
    #get indexes for unlabeled data
    for i,v in pos_vals.items():
        if i not in rand_pos_idx:
            unlab_idx.append(i)
    for i,v in neg_vals.items():
        if i not in rand_neg_idx:
            unlab_idx.append(i)
    unlab_data = semi_train_x[unlab_idx]
    new_train_x = semi_train_x[all_indexes]
    new_train_y = semi_train_y.values[all_indexes]
    while len(unlab_idx)-1 != -1:
        params_2 = {'C': [0.01, 0.1, 1, 10, 100]}
        svm_2 = LinearSVC(dual = False, max_iter=100000)
        #perform 5 fold cross validation
        cv_2 = GridSearchCV(svm_2, param_grid=params_2, cv=5, n_jobs= -1, error_score='raise')
        cv_2.fit(new_train_x, new_train_y.ravel())
        best_penalty = cv_2.best_params_['C']
        print('The penalty parameter is: '+str(best_penalty))
        semi_svm = LinearSVC(penalty='l1', C=best_penalty, dual = False, max_iter=100000)
        semi_svm.fit(new_train_x, new_train_y.ravel())
        # Calculating distance from decision boundary of unlabelled data
        decision_bound = cv_2.decision_function(unlab_data)
        w_norm = np.linalg.norm(cv_2.best_estimator_.coef_)
        distance = np.abs(decision_bound) / w_norm
        farthest_dist = np.argmax(distance)
        farthest_dp = unlab_idx[farthest_dist] 
        farthest_dp_unlab = semi_train_x[farthest_dp]        
        #let SVM label it and add it to the labeled data 
        pred_lab = semi_svm.predict(farthest_dp_unlab.reshape(1,-1))
        new_train_x = np.concatenate((new_train_x, farthest_dp_unlab.reshape(1,-1)), axis=0)
        new_train_y = np.concatenate((new_train_y, pred_lab.reshape(1,-1)), axis=0)
        unlab_idx = np.delete(unlab_idx, farthest_dist)
        unlab_data = semi_train_x[unlab_idx]
        if len(unlab_idx) == 0:
            cv_pred_test_2 = semi_svm.predict(semi_test_x)
            print('The final SVM test accuracy is: '+str(accuracy_score(semi_test_y, cv_pred_test_2)))
            cv_pred_train_2 = semi_svm.predict(new_train_x)
            acc_score_test_2 = accuracy_score(semi_test_y, cv_pred_test_2)
            acc_score_train_2 = accuracy_score(new_train_y, cv_pred_train_2)
            acc_list_test_2.append(acc_score_test_2)
            acc_list_train_2.append(acc_score_train_2)
            train_class_rep_2 = classification_report(new_train_y, cv_pred_train_2, output_dict=True)
            test_class_rep_2 = classification_report(semi_test_y, cv_pred_test_2, output_dict=True)
            precision_list_train_2.append(train_class_rep_2['weighted avg']['precision'])
            precision_list_test_2.append(test_class_rep_2['weighted avg']['precision'])
            recall_list_train_2.append(train_class_rep_2['weighted avg']['recall'])
            recall_list_test_2.append(test_class_rep_2['weighted avg']['recall'])
            f1_list_train_2.append(train_class_rep_2['weighted avg']['f1-score'])
            f1_list_test_2.append(test_class_rep_2['weighted avg']['f1-score'])
            auc_train_2 = roc_auc_score(new_train_y, cv_pred_train_2)
            auc_test_2 = roc_auc_score(semi_test_y, cv_pred_test_2)
            auc_list_train_2.append(auc_train_2)
            auc_list_test_2.append(auc_test_2)
            if m == 0:
                svm_fpr_test_2, svm_tpr_test_2, svm_thresh_test_2 = roc_curve(semi_test_y, cv_pred_test_2)
                svm_auc_test_2 = metrics.auc(svm_fpr_test_2, svm_tpr_test_2)
                svm_test_roc_2 = metrics.RocCurveDisplay(fpr=svm_fpr_test_2, tpr=svm_tpr_test_2, roc_auc=svm_auc_test_2)
                fig, ax = plt.subplots()
                svm_test_roc_2.plot(ax=ax, label='ROC Curve for Semi-Supervised SVM (Test Data)')
                plt.title('ROC Curve for Semi-Supervised SVM')
                svm_fpr_train_2, svm_tpr_train_2, svm_thresh_train_2 = roc_curve(new_train_y, cv_pred_train_2)
                svm_auc_train_2 = metrics.auc(svm_fpr_train_2, svm_tpr_train_2)
                svm_train_roc_2 = metrics.RocCurveDisplay(fpr=svm_fpr_train_2, tpr=svm_tpr_train_2, roc_auc=svm_auc_train_2)
                svm_train_roc_2.plot(ax=ax, label='ROC Curve for Semi-Supervised SVM (Train Data)')
                plt.legend()
                plt.show()
                conf_mat_test_2 = confusion_matrix(semi_test_y, cv_pred_test_2)
                print('The confusion matrix for testing data is:\n'+str(conf_mat_test_2))
                conf_mat_train_2 = confusion_matrix(new_train_y, cv_pred_train_2)
                print('The confusion matrix for training data is:\n'+str(conf_mat_train_2))
print('The average accuracy of the training data is: '+str(np.mean(acc_list_train_2)))
print('The average accuracy of the testing data is: '+str(np.mean(acc_list_test_2)))
print('The average precision of the training data is: '+str(np.mean(precision_list_train_2)))
print('The average precision of the testing data is: '+str(np.mean(precision_list_test_2)))
print('The average recall of the training data is: '+str(np.mean(recall_list_train_2)))
print('The average recall of the testing data is: '+str(np.mean(recall_list_test_2)))
print('The average f1-score of the training data is: '+str(np.mean(f1_list_train_2)))
print('The average f1-score of the testing data is: '+str(np.mean(f1_list_test_2)))
print('The average AUC of the training data is: '+str(np.mean(auc_list_train_2)))
print('The average AUC of the testing data is: '+str(np.mean(auc_list_test_2)))
```

#### (iv) Spectral Clustering

RESEARCH: 
-In short, the idea behind spectral clustering is to represent the data as a graph. In the graph there are nodes and edges which represent data points and similarity between the data points respectively.
-There are many advantages of spectral clustering over kmeans clustering, a significant one is that it has the ability to non-linearly separable data and can find clusters with complex shapes.
-A potential drawback to spectral clustering is that, in the context of large datasets, it may be computationally expensive.

### (b) Repeat each of the following two procedures 50 times

#### (i) Passive

```python
avg_fifty_err = []
all_errors_df_passive = pd.DataFrame(columns=list(range(1, 901, 10)),index=list(range(1,51))) 

for m in range(50):
    print('\nRUN #'+str(m+1))
    rand_pool_list = []
    deletion_df = pd.concat([norm_df_train_x, rand_train_set.iloc[:,4]], axis =1)
    passive_df = pd.DataFrame()
    passive_x_test = norm_df_test_x
    passive_y_test = rand_test_set.iloc[:,4]
    avg_test_err = []
    strat = StratifiedKFold(n_splits=3, shuffle=True)
    while len(passive_df.index) != 900:
        print('Pool #'+str((len(passive_df.index)//10)+1))
        rand_pool = deletion_df.sample(n=10)
        while True: #make sure that the value counts are 
            value_counts = rand_pool.iloc[:, 4].value_counts()
            if 1 in value_counts.index and value_counts[1] >= 3:
                break
            rand_pool = resample(deletion_df, n_samples=10)
            if 0 in value_counts.index and value_counts[0] >= 3:
                break
            rand_pool = resample(deletion_df, n_samples=10)
        passive_df = passive_df.append(rand_pool)
        deletion_df = deletion_df.drop(rand_pool.index)
        passive_x_train = passive_df.iloc[:,0:4]
        passive_y_train = passive_df.iloc[:,4]
        new_svm = LinearSVC(dual = False, max_iter=100000, class_weight='balanced')
        new_params = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
        new_cv = GridSearchCV(new_svm, param_grid=new_params, cv=strat, n_jobs= -1, error_score='raise')
        new_cv.fit(passive_x_train, passive_y_train)
        new_best_param = new_cv.best_params_['C']
        new_svm_2 = LinearSVC(penalty='l1', C=new_best_param, dual=False, class_weight='balanced', max_iter=100000)
        print('The best penalty parameter is: '+str(new_best_param))
        new_svm_2.fit(passive_x_train, passive_y_train)
        passive_pred = new_svm_2.predict(passive_x_test)
        passive_acc = accuracy_score(passive_y_test, passive_pred)
        test_error = 1 - passive_acc
        avg_test_err.append(test_error)
        print('The test error for SVM #'+str((len(passive_df.index)//10))+' is: '+str(test_error)+'\n')
    all_errors_df_passive.loc[m+1] = avg_test_err     
```

#### (ii) Active

```python
avg_fifty_err_2 = []
all_errors_df_active = pd.DataFrame(columns=list(range(1, 901, 10)),index=list(range(1,51))) 
for m in range(50):
    rand_pool_list_2 = []
    deletion_df_2 = rand_train_set
    active_df = pd.DataFrame()
    all_datapoints_x = norm_df_train_x
    all_datapoints_y = rand_train_set.iloc[:,4]
    all_datapoints_test_x = norm_df_test_x
    all_datapoints_test_y = rand_test_set.iloc[:,4]
    avg_test_err_2 = []
    strat = StratifiedKFold(n_splits=3)
    print('RUN #'+str(m+1))
    while len(active_df.index) != 900:
        print('Pool #'+str((len(active_df.index)//10)+1))
        print('deletion_df size: '+str(len(deletion_df_2)))
        rand_pool_2 = deletion_df_2.sample(n=10)
        while True: #make sure that the value counts are 
            value_counts = rand_pool_2.iloc[:, 4].value_counts()
            if 1 in value_counts.index and value_counts[1] >= 3:
                break
            rand_pool_2 = resample(deletion_df_2, n_samples=10)
            if 0 in value_counts.index and value_counts[0] >= 3:
                break
            rand_pool_2 = resample(deletion_df_2, n_samples=10)
        active_df = active_df.append(rand_pool_2)
        deletion_df_2 = deletion_df_2.drop(rand_pool_2.index)
        active_x_train = active_df.iloc[:,0:4]
        active_y_train = active_df.iloc[:,4]
        new_svm_3 = LinearSVC(dual = False, max_iter=10000)
        new_params_2 = {'C': [0.01, 0.1, 1, 10, 100]}
        new_cv_2 = GridSearchCV(new_svm_3, param_grid=new_params_2, cv=strat, n_jobs= -1)
        new_cv_2.fit(active_x_train, active_y_train)
        new_best_param_2 = new_cv_2.best_params_['C']
        new_svm_4 = LinearSVC(penalty='l1', C=new_best_param_2, dual=False, max_iter=100000)
        new_svm_4.fit(active_x_train, active_y_train)
        print('The best penalty parameter is: '+str(new_best_param))
        active_dist = abs(new_svm_4.decision_function(all_datapoints_x) / (np.linalg.norm(new_svm_4.coef_) + 1e-8))
        sort_dist = np.argsort(active_dist)
        ten_closest_idx = sort_dist[:10]
        ten_closest_dp = all_datapoints_x.index[ten_closest_idx]
        ten_closest_df = all_datapoints_x.loc[ten_closest_dp]
        active_df.append(ten_closest_df)
        all_datapoints_x = all_datapoints_x.drop(ten_closest_df.index)
        all_datapoints_y = all_datapoints_y.drop(ten_closest_df.index)
        active_pred = new_svm_4.predict(all_datapoints_test_x)
        active_acc = accuracy_score(all_datapoints_test_y, active_pred)
        active_test_err = 1- active_acc
        avg_test_err_2.append(active_test_err)
        print('The test error for SVM #'+str((len(active_df.index)//10))+' is: '+str(active_test_err)+'\n')
    all_errors_df_active.loc[m+1] = avg_test_err_2     
```
